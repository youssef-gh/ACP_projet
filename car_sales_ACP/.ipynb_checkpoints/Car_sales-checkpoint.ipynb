Cette dataset, Décrit 30 model de voitures selon les 12 variables suivantes

- Sales_in_thousands : Ventes en milliers

- year_resale_value : valeur de revente annuelle

- Price_in_thousands : Prix en milliers

- Engine_size : La taille du moteur

- Horsepower : Puissance des chevaux

- Wheelbase : Empattement

- Width : Largeur

- Length : Longueur

- Curb_weight : Poids à vide

- Fuel_capacity : Capacité du carburant

- Fuel_efficiency : La consommation du carburant

- Power_perf_factor : Facteur de performance de puissance

```python
import pandas as pd
import numpy as np

#chargement de notre fichier csv
X = pd.read_csv("car_sales.csv",delimiter=',').set_index("Name,Model").fillna(0)
X.drop(["Manufacturer","Latest_Launch","Vehicle_type","Model"], axis=1, inplace=True)

dataset=X.head(30)
dataset
```

```python
#dimension de notre matrice
print(dataset.shape) 

#nombre d'observations (lignes)
n = dataset.shape[0]

#nombre de variables (colonnes)
p = dataset.shape[1]
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
```

<font color="blue"><b>Centrage et réduction de notre matrice<b> <font>

```python
sc = StandardScaler()

#La matrice Z est centrée et réduite

Z = sc.fit_transform(dataset) 
print(Z)
```

<FONT COLOR="BLUE"><B>On peut bien s'assurer que les moyennes des colonnes sont nulles <B> <FONT> 

```python
#moyenne
print(np.mean(Z,axis=0))
```

<FONT COLOR="BLUE"><B>Et que les écarts-type sont égales à 1.<B> <FONT> 

```python
#écart-type
print(np.std(Z,axis=0))
```

<FONT COLOR="GREEN"><B><I>1 . Calculs d'ACP<I><B> <FONT> 

```python
#importation de PCA de la librairie sklearn
from sklearn.decomposition import PCA


acp = PCA(svd_solver='full')
```

<FONT COLOR="BLUE"><B>Calcule de la matrice des corrélations <B> <FONT> 

```python
coord = acp.fit_transform(Z)
print(coord) 
```

<FONT COLOR="GREEN"><B><I>2 . Valeurs propres et éboulis des valeurs propres!<I><B> <FONT> 

<FONT COLOR="blue"><B>Calcul des valeurs propres λ<B> <FONT> 

```python

valeurs_propres=acp.explained_variance_

print(valeurs_propres)
```

<FONT COLOR="blue"><B>Calcule des quatités d'information a partir des valeurs propres
<B> <FONT> 

```python
print(acp.explained_variance_ratio_)
```

<FONT COLOR="green">On remarque que la première composante contient 59.59% d'information
    <FONT> 

<FONT COLOR="blue"><B>L'éboulis des valeurs propres  <B> <FONT> 

L'éboulis est un diagramme qui décrit le pourcentage d'inertie totale associé a chaque axe

```python

import matplotlib.pyplot as plt

#scree plot
figure = plt.figure(figsize=(8,8))
plt.plot(np.arange(1,len(dataset.columns)+1),valeurs_propres) 
plt.title("éboulis des valeurs propres") 
plt.ylabel("valeurs propres") 
plt.xlabel("nombre de composantes principales") 
plt.show()
```

On séléctionne les axes avant le décrochement

 <FONT COLOR="green"><B>4 . Représentation des individus – Outils pour l’interprétation
     <B> <FONT>

<B> <FONT COLOR="blue">a .Coordonnées factorielles:<FONT > <FONT COLOR="black">
Les coordonnées factorielles (Fik) des individus ont été collectées dans la variable coord . Nous les positionnons dans le premier , le deuxieme  plan factoriel avec leurs labels pour situer et comprendre les proximités entre les individus.! <FONT ><B>

```python
pd.DataFrame(coord, columns=["F1", "F2", "F3", "F4","F5", "F6", "F7", "F8", "F9", "F10", "F11", "F12"]).loc[:, ["F1", "F2"]]
```

<FONT COLOR="black"><B>     Représentation des individus dans le plan factoriel
<FONT COLOR="green"><B>

<FONT COLOR="blue"><B>b . Qualité de représentation – Les COS² (cosinus carré). <FONT><B>

```python

#contribution des individus dans l'inertie totale 
#somme de Zij (d^2 = sum(Zij))
di = np.sum(Z**2,axis=1) # les carrés des distances à l’origine des individus (zij)

print(pd.DataFrame({'Name':dataset.index,'dstance':di}))
#di: les carrés des distances à l’origine des individus
```

<FONT COLOR="blue"><B>commentaire<FONT COLOR="black"><B>

```python

#qualité de représentation des individus - COS2
# cos^2 = Fik^2/di^2

cos2 = coord**2 
for j in range(p):
    cos2[:,j] = cos2[:,j]/di

print(pd.DataFrame({'id':dataset.index,'COS2_1':cos2[:,0],'COS2_2':cos2[:,1]}))
```

<font color="blue"><b>pour chaque pays, la somme des COS² sur l’ensemble des
facteurs est égale à 1.<font><b>


```python

#vérifions la théorie - somme en ligne des cos2 = 1
print(np.sum(cos2,axis=1))
```

<FONT COLOR="blue"><B>c . Contribution des individus aux axes (CTR) :
Elles permettent de déterminer les individus qui pèsent le plus dans la définition de chaque facteur.!<FONT><B>

```python

#contributions aux axes
#CTR = Fik^2/n*𝜆i
ctr = coord**2
for j in range(p):
    ctr[:,j] = ctr[:,j]/(n*valeurs_propres[j])

print(pd.DataFrame({'id':dataset.index,'CTR_1':ctr[:,0],'CTR_2':ctr[:,1]}))
```

 <FONT COLOR="blue"><B>Cadillac Escalade et Chevrolet Metro qui est un  déterminant pour le premier axe ; pour le second, nous avons Chevrolet Corvette et Chevrolet Impala <FONT COLOR="blue"><B>

```python

#vérifions la théorie "Les sommes en ligne sont égales à 1"
print(np.sum(ctr,axis=0))
```

<b><font color="green">5 . Représentation des variables – Outils pour l’aide à l’interprétation!<font><b>

<font color="blue"><b>Nous avons besoin des vecteurs propres pour l’analyse des variables.<br>Ils sont fournis par le
champ .components_ : pour calculer les vecteurs proprs<font><b>

```python
#le champ components_ de l'objet ACP
print(acp.components_)
```

<b><font color="blue">calcul de la corrélation <font><b>

```python

#racine carrée des valeurs propres
sqrt_val_propr = np.sqrt(valeurs_propres)

#les facteurs sont en ligne, les variables en colonne. 
#Nous devons en tenir compte pour obtenir les corrélations (variables x facteurs, 𝑟𝑗𝑘)
#en les multipliant par la racine carrée des valeurs propres :



#corrélation des variables avec les axes
correlation = np.zeros((p,p))

for k in range(p):
    correlation[:,k] = acp.components_[k,:] * sqrt_val_propr[k]

#afficher la matrice des corrélations variables x facteurs
#Les variables sont maintenant en ligne, les facteurs en colonne :
pd.DataFrame(correlation, index=dataset.columns)
```

```python

#on affiche la correlation pour les 2 premiers axes
print(pd.DataFrame({'id':dataset.columns,'COR_1':correlation[:,0],'COR_2':correlation[:,1]}))
```

<b> <font color="blue">le cercle des corrélations <font><b>

```python

#cercle des corrélations
fig, axes = plt.subplots(figsize=(8,8)) 
axes.set_xlim(-1,1)
axes.set_ylim(-1,1)

#affichage des étiquettes (noms des variables)
for j in range(p): 
    plt.scatter(correlation[j,0],correlation[j,1])
    plt.annotate(dataset.columns[j],(correlation[j,0],correlation[j,1]))

#ajouter les axes
plt.plot([-1,1],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-1,1],color='silver',linestyle='-',linewidth=1)

#ajouter un cercle
cercle = plt.Circle((0,0),1,color='blue',fill=False) 
axes.add_artist(cercle)

#affichage
plt.show()

#Les signes sont opposés Mais ce n’est pas un problème, ce
#sont les concomitances et oppositions qui comptent.
```

<font color="blue"><b>comentaire <b><font>

Plus on se deplace vers la gauche , on trouve les "models" qui ont un petit nombre horsepower,engine_size , Fuel_capacity, width , power_perf_factor C'est ce qu'on trouve dans la matrice des corrélations (ces variables sont corrélés négativement avec le 1er axe).

Plus on monte vers le haut, on trouve les models qui ont un grand nombre de year_resale_value, price_in_thousands, power_perf_factor,. C'est ce qu'on trouve dans la matrice des corrélations (ces variables sont corrélés positivement  le 2eme axe)

Plus on descend vers le bas, on trouve les models qui ont un grand nombre de length, curb_weight, wheelbase 'est ce qu'on trouve dans la matrice des corrélations (ces variables sont corrélés négativement  le 2eme axe)

<b><font color="blue">a  . Qualité de représentation des variables (COS²).<font color="blue"><font color="black"> On peut calculer la qualité de
représentation des variables en montant la corrélation au carré : 𝐶𝑂𝑆^2(jk) = 𝑟^2(jk) <b><font>

```python

#cosinus carré des variables
cos2var = correlation**2
print(pd.DataFrame({'id':dataset.columns,'COS2_1':cos2var[:,0],'COS2_2':cos2var[:,1]}))
```

<b><font color="blue">La somme en ligne :la somme des COS² d’une variable sur l’ensemble des facteurs est égale à 1 ).<b><font color="blue">


<b><font color="blue">b  . Contribution des variables aux axes (CTR):<font color="blue"><font color="black"> La contribution est également basée sur le
carré de la corrélation, mais relativisée par l’importance de l’axe <br> CTRjk = rjk/𝜆k
 <b><font>

```python

#contributions
ctrvar = cos2var

for k in range(p):
    ctrvar[:,k] = ctrvar[:,k]/valeurs_propres[k]

#on n'affiche que pour les 2 premiers axes
print(pd.DataFrame({'id':dataset.columns,'CTR_1':ctrvar[:,0],'CTR_2':ctrvar[:,1]}))
```

<font color="blue">Les sommes en colonnes sont égales à 1 cette fois-ci.<font color="blue">
